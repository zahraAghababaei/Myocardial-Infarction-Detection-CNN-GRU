#### Now try to combine all csv files to one csv.

```python

control_path = './peakcontrol/'
mi_path = './peakmi/'
output_control_csv_file = 'combined_control1.csv'
output_mi_csv_file = 'combined_mi1.csv'
```

Read all control csv files to find out which one has longer signal

```python
row_count = 0
max_control_row_count = 0

for filename in os.listdir(control_path):
    if filename.endswith('.csv'):
        with open(os.path.join(control_path, filename), 'r') as file:
            reader = csv.reader(file)
            row_count = len(list(reader))
            if max_control_row_count < row_count:
                max_control_row_count = row_count
                print(f'{filename}:{max_control_row_count}')
            #max_row_count = max(max_row_count, row_count)

print(max_control_row_count)
```

Now combine all together

```python
max_control_row_count = 500

with open(output_control_csv_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',', escapechar=' ', quoting=csv.QUOTE_NONE)
    for filename in os.listdir(control_path):
        if filename.endswith('.csv'):
            with open(os.path.join(control_path, filename), 'r') as file:
                print(filename)
                reader = csv.reader(file)
                rows = []
                for row in reader:
                    rows.append('{:.18e}'.format(float(row[0])))
                row_len = len(rows)  # Get the number of columns in this file
                if max_control_row_count > row_len:
                    rows.extend([0] * (max_control_row_count - row_len))   
                if max_control_row_count < row_len:
                    rows = rows[:max_control_row_count]
                rows.append('{:.18e}'.format(0.0)) 
                writer.writerow(rows[:])

            row_count = 0  # Reset the row count for the next file

csvfile.close()
```
